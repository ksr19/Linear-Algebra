__1.__ Исследовать на линейную зависимость:

$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$

По определению векторы $u_{1},u_{2}, ...,u_{n}$ являются _линейно зависимыми_, если найдутся такие числа $\lambda_{1}, \lambda_{2}, ..., \lambda_{n}$, что хотя бы одно из них отлично от нуля и 
 
 
### $$\lambda_{1}u_{1}+\lambda_{2}u_{2}+...+\lambda_{n}u_{n}=0.$$
В данном случае предыдущее выражение можно записать следующим образом:
### $$f_{1}(x) + f_{2}(x) - f_{3}(x) + f_{4}(x) = e^x + 1 - x - 1 + x - e^x = 0$$
То есть в данном случае $\lambda_{1} = \lambda_{2} = \lambda_{4} = 1; \lambda_{3} = -1$. Это значит, что векторы являются ___линейно зависимыми___.

__2.__ Исследовать на линейную зависимость:
$$f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}$$

По определению векторы $u_{1},u_{2}, ...,u_{n}$ являются _линейно зависимыми_, если найдутся такие числа $\lambda_{1}, \lambda_{2}, ..., \lambda_{n}$, что хотя бы одно из них отлично от нуля и 
 
 
### $$\lambda_{1}u_{1}+\lambda_{2}u_{2}+...+\lambda_{n}u_{n}=0.$$
В данном случае предыдущее выражение можно записать следующим образом:
### $$\frac{1}{2}\cdot f_{1}(x) + 2\cdot f_{2}(x) + f_{3}(x) - f_{4}(x) = 1 + 2x + x^2 - (x+1)^2 = 0$$
То есть в данном случае $\lambda_{1} = \frac{1}{2}; \lambda_{2} = 2; \lambda_{3} = 1; \lambda_{4} = -1$. Это значит, что векторы являются ___линейно зависимыми___.

__5.__ Установить, является ли линейным подпространством:

а) совокупность всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю;
    
б) все векторы, являющиеся линейными комбинациями данных векторов $\{u_{1}, u_{2}, \ldots, u_{n}\}$.

а) 
По определению подмножество $L$ линейного пространства $V$ является его подпространством тогда и только тогда, когда для любых элементов $u, v\in L$ и любого $\alpha \in \mathbb{R}$ выполняются условия:

$$1)\ u+v\in L;$$

$$2)\ \alpha\cdot u \in L.$$

Для рассматриваемой в данном пункте совокупности векторов нарушается первое условие. Рассмотрим, например, векторы $u = (0, 1, 1)$ и $v = (1, 0, 1)$. Тогда $u + v = (1, 1, 2) \notin L$. То есть совокупность всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю ___не является___ линейным подпространством.

б) По определению подмножество $L$ линейного пространства $V$ является его подпространством тогда и только тогда, когда для любых элементов $u, v\in L$ и любого $\alpha \in \mathbb{R}$ выполняются условия:

$$1)\ u+v\in L;$$

$$2)\ \alpha\cdot u \in L.$$

Возьмем векторы $a_1 = \lambda_1u_1+\lambda_2u_2+...+\lambda_nu_n$ и $a_2 = \mu_1u_1+\mu_2u_2+...+\mu_nu_n$ и проверим выполнение представленных выше условий: <br>
1. Пользуюсь свойством дистрибутивности относительно сложения скаляров, получаем:
### $$ a_1 + a_2 = \lambda_1u_1+\mu_1u_1 + \lambda_2u_2+\mu_2u_2 + ...+\lambda_nu_n+\mu_nu_n$$
### $$a_1+a_2 = (\lambda_1+\mu_1)u_1 +(\lambda_2+\mu_2)u_2+...+(\lambda_n+\mu_n)u_n = \phi_1u_1+\phi_2u_2+...+\phi_nu_n\in L$$
2. Пользуюсь свойством ассоциативности операции умножения на скаляр, получаем:
### $$ \alpha\cdot a_1 = \alpha \cdot ( \lambda_1u_1+\lambda_2u_2+...+\lambda_nu_n)$$
### $$ \alpha\cdot a_1 = \alpha\lambda_1u_1+\alpha\lambda_2u_2+...+\alpha\lambda_nu_n = \rho_1u_1+\rho_2u_2+...+\rho_nu_n\in L$$
Оба условия выполняются, а это значит, что все векторы, являющиеся линейными комбинациями данных векторов $\{u_{1}, u_{2}, \ldots, u_{n}\}$ образуют линейное подпространство.

__3.__ Будет ли линейное пространство евклидовым, если за скалярное произведение принять:<br>
а) произведение длин векторов;<br>
б) утроенное обычное скалярное произведение векторов?<br>

а) В таком случае $\langle x, y \rangle = \left \|x\right \| \cdot \left \|y\right \|$. Согласно определению, линейное пространство над полем вещественных чисел называется _евклидовым пространством_, если на нем введено правило, согласно которому каждой упорядоченной паре векторов $x$ и $y$ поставлено в соответствие вещественное число, называемое _скалярным произведением_ и обозначаемое символом $\langle x, y \rangle$.

При этом указанное правило должно подчиняться четырем аксиомам:<br>
1) $\langle x, y \rangle=\langle y, x \rangle;$<br>
2) $\langle \lambda x, y \rangle=\lambda\langle x, y \rangle;$<br>
3) $\langle x_1 + x_2, y \rangle=\langle x_1, y \rangle+\langle x_2, y \rangle;$<br>
4) $\langle x, x \rangle\geq0$, причем $\langle x, x \rangle=0\Leftrightarrow x=0.$<br>

Рассмотрим третий пункт представленной аксиоматики и перепишем его с учетом заданного скалярного произведения:
#### $\langle x_1 + x_2, y \rangle=\left \|x_1+x_2\right \| \cdot \left \|y\right \|$
#### $\langle x_1, y \rangle+\langle x_2, y \rangle = \left \|x_1\right \| \cdot \left \|y\right \| + \left \|x_2\right \| \cdot \left \|y\right \|$
Однако в общем случае $\left \|x_1+x_2\right \| \neq \left \|x_1\right \| + \left \|x_2\right \|$. Например, если мы пользуемся $\ell_{2}$ нормой и рассматриваем векторы $x_1 = (4, 2, 4); x_2 = (0, 0, 1)$. Тогда $\left \|x_1+x_2\right \| = \sqrt{45}$, в то время как $\left \|x_1\right \| + \left \|x_2\right \| = 6 + 1 =7$. <br>
Таким образом линейное пространство, в котором скалярное произведение векторов задается произведением длин этих векторов, ___не является евклидовым___.

б) В таком случае $\langle x, y \rangle_{new} = 3 \langle x, y \rangle_0=3\cdot\left \|x\right \| \cdot \left \|y\right \| \cdot cos\alpha,$ где $\alpha$ - угол между векторами. Согласно определению, линейное пространство над полем вещественных чисел называется _евклидовым пространством_, если на нем введено правило, согласно которому каждой упорядоченной паре векторов $x$ и $y$ поставлено в соответствие вещественное число, называемое _скалярным произведением_ и обозначаемое символом $\langle x, y \rangle$.

При этом указанное правило должно подчиняться четырем аксиомам:<br>
1) $\langle x, y \rangle=\langle y, x \rangle;$<br>
2) $\langle \lambda x, y \rangle=\lambda\langle x, y \rangle;$<br>
3) $\langle x_1 + x_2, y \rangle=\langle x_1, y \rangle+\langle x_2, y \rangle;$<br>
4) $\langle x, x \rangle\geq0$, причем $\langle x, x \rangle=0\Leftrightarrow x=0.$<br>

Проверим все пункты по порядку, пользуясь тем фактом, что скалярного произведения $\langle x, y \rangle_0$ указанные условия соблюдаются:
1) $\langle x, y \rangle_{new} = 3 \langle x, y \rangle_0 = 3\langle y, x \rangle_0 = \langle y, x \rangle_{new};$ <br>
2) $\langle \lambda x, y \rangle_{new}=3\langle \lambda x, y \rangle_0 =\lambda\cdot3\langle x, y \rangle_0=\lambda\langle x, y \rangle_{new};$<br>
3) $\langle x_1 + x_2, y \rangle_{new}=3\langle x_1 + x_2, y \rangle_0 = 3\langle x_1, y \rangle_0+3\langle x_2, y \rangle_0 = \langle x_1, y \rangle_{new} + \langle x_2, y \rangle_{new};$<br>
4) $\langle x, x \rangle_{new}\geq0$, причем $\langle x, x \rangle_{new}=0\Leftrightarrow x=0.$<br>

Все условия для скалярного произведения заданного таким образом соблюдаются, а это значит, что линейное пространство ___является евклидовым___.
